### 4) PieChart

```python

df=pd.read_csv('Downloads/courses.csv')
df.head()
```

```python
print('data dimensions:', df.shape)
```

```python
df_continents = df.groupby('course_title', axis=0).sum()

# note: the output of the groupby method is a `groupby' object. 
# we can not use it further until we apply a function (eg .sum())
print(type(df.groupby('course_title', axis=0)))

df1=df_continents.head()
df1
```

```python
explode_list = [0, 0, 0, 0, 0]
df1['price'].plot(kind='pie',
                            figsize=(5, 6),
                            autopct='%1.1f%%',
                            startangle=90,  
                            shadow=True,
                             pctdistance=1.15,
                            labels=None,
                           # add shadow      
                            )

plt.title('Cources')
plt.axis('equal') # Sets the pie chart to look like a circle.

plt.show()
```

### 5) Scatter Plot

```python
df1.plot(kind='scatter', x='price', y='num_lectures', figsize=(10, 6), color='skyblue')
plt.title('Cources')
plt.xlabel('Country-code')
plt.ylabel('Region-code')
plt.show()
```

### 6) LineChart

### 7) Wordcloud

```python
from wordcloud import WordCloud, STOPWORDS

print ('Wordcloud is installed and imported!')
```

```python
text_file = open('Downloads/Sp.txt', 'r').read()
    
print ('File downloaded and saved!')
```

```python
stopwords = set(STOPWORDS)
```

```python
wc = WordCloud(
    background_color='white',
    max_words=2000,
    stopwords=stopwords
)

# generate the word cloud
wc.generate(text_file)
```

```python
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

### 8) Catplot

```python
dt=df['content_duration'].str.split('')
df['hour']=dt.str.get(1)
df['hour']=df['hour'].astype(int)
```

```python
sns.catplot(x="hour",y="subject",kind='bar',data=df);
```

```python

plt.subplot(142)
plt.title('Count of Paid Vs. Free Courses',fontsize = 15)
sns.countplot(df['is_paid'])
```

```python
#this displays the distribution of the number of subscribers
plt.figure(figsize=(10,8))
plt.style.use('ggplot')
sns.distplot(np.log(df['num_subscribers']+1),color='blue')
plt.xlabel('Number of Subscribers')
plt.title('Normal distribution of the number of subscribers')
```

```python
#this displays the Corr. between Number of Subscribers & price
plt.figure(figsize=(10,8))
plt.title('Corr. between Number of Subscribers and Price',fontsize = 20)
sns.regplot(y=df['num_subscribers'],x=df['price'])
plt.style.use('ggplot')
plt.xticks(rotation = 90)
```
